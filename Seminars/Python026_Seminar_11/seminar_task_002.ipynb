Дана функция f(x) = 5x^2 + 10x - 30

1. Определить корни                                     

2. Найти интервалы, на которых функция возрастает       

3. Найти интервалы, на которых функция убывает          

4. Построить график                                     

5. Вычислить вершину                                    

6. Определить промежутки, на котором f > 0              

7. Определить промежутки, на котором f < 0              


**Задаем функцию**

```python
import math
from sympy import *
from sympy.plotting import plot as plt
init_printing()

x = Symbol('x')
fx = 5*x**2 + 10*x - 30
```

**1. Определить корни:**

```python
x1, x2 = solve(fx)
x1 = round(x1.evalf(), 2)
x2 = round(x2.evalf(), 2)
print(f'первый корень = {x1}\nвторой корень = {x2}')
```

**2. Найти интервалы, на которых функция возрастает**

```python
interval_1=solve(diff(fx)>0)
print(f'Функция {fx} возрастает на интервале(ах): {interval_1}')
interval_1
```

**3. Найти интервалы, на которых функция убывает**

```python
interval_2=solve(diff(fx)<0)
print(f'Функция {fx} убывает на интервале(ах): {interval_2}')
interval_2
```

**4. Построить график**

```python
p = plt(fx, (x, -10, 10))
```

**5. Вычислить вершину**

```python
xv = solve(diff(fx))[0]
yv = 5*xv**2 + 10*xv - 30
xv, yv
```

**6. Определить промежутки, на котором f > 0**

```python
solve(fx > 0)
# print(solve(fx>0))
```

**7. Определить промежутки, на котором f < 0**

```python
solve(fx<0)
# print(solve(fx<0))
```
